By running this code it removes html tags and separates user and odinschool chats into different columns

with lambda functions and list comprehension to make it work for large files  

```python
import pandas as pd
import re

df = pd.read_excel('/content/Threads__4.xlsx')
df=df['Thread Content']

df1= pd.read_excel('/content/output2.xlsx')

CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
CLEANR1 = re.compile("wrote ----")
CLEANR2 = re.compile("wrote:")

x = lambda raw_html: re.sub(CLEANR, '', raw_html)
y = lambda x: re.sub(CLEANR1, "\nuser:", x)
z = lambda y: re.sub(CLEANR2, "\nodinschool:" , y)
p = lambda z: re.split(r'\n(?=\w+:)', 'odinschool:'+z)

df = df.apply(lambda val: p(z(y(x(str(val))))))

odinschool = df.apply(lambda x: [elem for elem in x if elem.startswith('odinschool:')])
user = df.apply(lambda x: [elem for elem in x if elem.startswith('user:')])

df1['Odinschool'] = odinschool
df1['User']= user
df1.to_excel('/content/output2.xlsx')
```

```python
import re
import pandas as pd
import string

df = pd.read_excel('/content/Threads__4.xlsx')
raw_html=df['Thread Content']
raw_lst=[]
raw_lst.append(raw_html)
raw_html= str(raw_html)

CLEANR = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});')
CLEANR1= re.compile("wrote ----")
CLEANR2= re.compile("wrote:")

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  cleantext= re.sub(CLEANR2,("\033[1m" + "\nuser:" + "\033[0m"), cleantext)
  cleantext= re.sub(CLEANR1,("\033[1m" + "\nodinschool:" + "\033[0m"),cleantext)
  return "\033[1m" + "\nodinschool:" + "\033[0m"+cleantext

cleanhtml=cleanhtml(raw_html)
print(cleanhtml)

def splitting(cleanhtml):
  messages = re.split(r'\n(?=\w+:)', cleanhtml)
  odinschool=[]
  user=[0]
  for message in messages:
      if message.startswith('odinschool:'):
        odinschool.append(message)
      elif message.startswith('user:'):
        user.append(message)
  df['odinschool']=odinschool
  df['user']=user
  return df


print(splitting(cleanhtml))
df.to_csv('result1.csv')
```
